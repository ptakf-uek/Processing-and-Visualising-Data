```python
import pandas as pd

cities_in_poland = pd.read_csv("cities-in-poland.csv", sep=";")
cities_in_poland
```

* display last 8 rows of data with columns: City, Province, Population

```python
cities_in_poland.loc[:,["City", "Province", "Population"]].tail(8)
```

* calculate and add population density to the data frame (add a new column)

```python
cities_in_poland["Density"] = (
    cities_in_poland["Population"] / cities_in_poland["Areakm2"]
)

cities_in_poland
```

* display first 5 rows with columns: City, Density

```python
cities_in_poland.loc[:,["City", "Density"]].head(5)
```

* sort cities alphabetically and display the first 10 cities

```python
cities_in_poland.sort_values("City").head(10)
```

* display cities with at least half a million inhabitants (How many such cities are there in Poland?)

```python
cities_in_poland.query("Population >= 500000")#.shape[0]
```

* display 10 cities with the highest population

```python
cities_in_poland.sort_values("Population", ascending=False).head(10)
```

* display 10 cities with the least population

```python
cities_in_poland.sort_values("Population").head(10)
```

* display cities from the 'małopolskie' province

```python
cities_in_poland.query("Province == 'małopolskie'")
```

* display cities from the 'małopolskie' province and 'tarnowski' region

```python
cities_in_poland.query("Province == 'małopolskie' and Region == 'tarnowski'")
```

* display cities from the 'małopolskie' province and 'tarnowski' region, sorted by area, starting with the largest one

```python
cities_in_poland.query(
    "Province == 'małopolskie' and Region == 'tarnowski'"
).sort_values("Areakm2", ascending=False)
```
