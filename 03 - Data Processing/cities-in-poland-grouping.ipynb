```python
import pandas as pd

cities = pd.read_csv('cities-in-poland.csv', sep=';')
```

The list of Polish cities is available in the cities-in-poland.csv file. Calculate and display:

The number of cities in each province

```python
cities.loc[:,['Province', 'City']].groupby('Province').count()
```

The total number of inhabitants in each province

```python
cities.loc[:,['Province', 'Population']].groupby('Province').sum()
```

The total number of inhabitants in each of the regions in the 'opolskie' province

```python
cities.query("Province == 'opolskie'").loc[:,['Region', 'Population']] \
.groupby('Region').sum()
```

The total number of inhabitants in each of the regions in the 'opolskie' province; sort the values in descending order

```python
cities.query("Province == 'opolskie'").loc[:,['Region', 'Population']] \
.groupby('Region').sum().sort_values('Population', ascending=False)
```
