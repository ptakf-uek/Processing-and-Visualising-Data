The list below contains the tennis players who have won the most Grand Slam tournaments. Create a new notebook and then a chart presenting the data from the table. Remember to include the title and description of the chart axis.

Player         | Titles
---------------|-------
Novak Djokovic | 24
Rafael Nadal   | 22
Roger Federer  | 20
Pete Sampras   | 14


```python
import pandas as pd
import matplotlib

players = pd.DataFrame({"Player": ["Novak Djokovic", "Rafael Nadal",
                                   "Roger Federer", "Pete Sampras"],
                        "Titles": [24, 22, 20, 11]})

players.plot(kind="bar", title="Tennis players who have won the most GS tournaments",
             xlabel="Player", ylabel="Titles", x="Player")

```
